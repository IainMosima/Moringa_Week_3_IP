### Determinig the most used city/cities for three days

### Determinig the most used city/cities during business hours i.e between 9am and 5 pm

```python
# creating a new column called hours
#convert from string to datetime
final_2['DATETIME'] = pd.to_datetime(final_2['DATETIME'])
final_2['hour'] =  final_2['DATETIME'].dt.hour
final_2
```

```python
final_2.dtypes
```

```python
# Determing business hours
business_hours = final_2[(final_2['hour'] > 9) & (final_2['hour'] < 17)]
business_hours
```

### Determinig the most used city/cities during home hours i.e between 5:01 pm and 8:59 am.

```python
# First start by creating a dataset know as home_hours
home_hours = final_2[~(final_2['hour'] > 9) & (final_2['hour'] < 17)]
home_hours
```

```python
# Determing the maximum billing values
max_home_hours = home_hours['VALUE'].max()
most_used_home_hours = home_hours[(home_hours['VALUE'] == max_home_hours)]
most_used_home_hours
```

```python
# Determing the mean of the most used city during home hours
group = most_used_home_hours['VALUE'].groupby(most_used_home_hours['VILLES'])
group.mean()
```
